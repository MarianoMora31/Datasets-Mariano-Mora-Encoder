Dataset 1: ncr_ride_bookings.csv

Drop Location (categórica): indica dónde finaliza el viaje; es clave para analizar patrones de demanda según zonas geográficas.

Avg VTAT y Avg CTAT (numéricas): representan tiempos promedio de turnaround y cycle; útiles para evaluar eficiencia operativa y tiempos de espera.

Booking Value (numérica): refleja el valor económico del viaje; central para medir ingresos y analizar rentabilidad.

Ride Distance (numérica): distancia recorrida; importante para estudiar relación con valor, duración o satisfacción.

Driver Ratings y Customer Rating (numéricas): variables de calidad percibida; fundamentales para medir servicio y satisfacción.

Payment Method (categórica): permite segmentar usuarios según forma de pago, lo que aporta a decisiones financieras/comerciales.

😴 Dataset 2: Sleep_health_and_lifestyle_dataset.csv

Age y Gender (categóricas/numéricas): permiten segmentar hábitos de sueño y salud según demografía.

Sleep Duration y Quality of Sleep (numéricas): variables centrales para medir calidad del descanso.

Physical Activity Level y Stress Level (numéricas): indicadores de estilo de vida, influyen directamente en la salud.

BMI Category (categórica): clasifica peso corporal, útil para cruzar con patrones de sueño y actividad.

Blood Pressure, Heart Rate, Daily Steps (numéricas/texto): métricas fisiológicas que ayudan a vincular sueño con salud cardiovascular.

Sleep Disorder (categórica): variable dependiente clave para analizar factores asociados a trastornos del sueño.

❤️ Dataset 3: CVD_cleaned.csv

General_Health (categórica): autopercepción de salud, relevante como variable global de estado.

Exercise, Smoking_History, Alcohol_Consumption (categóricas/numéricas): factores de estilo de vida con alto impacto en riesgo cardiovascular.

Diabetes, Depression, Arthritis, Cancers (categóricas): comorbilidades importantes para el análisis de salud integral.

Height_(cm), Weight_(kg), BMI (numéricas): medidas antropométricas fundamentales para riesgos de CVD (cardiovascular disease).

Fruit_Consumption, Green_Vegetables_Consumption, FriedPotato_Consumption (numéricas): reflejan hábitos alimenticios, vinculados a la prevención o riesgo de CVD.

Heart_Disease (categórica): variable dependiente principal, clave para modelos predictivos o estudios de factores asociados.
