Dataset 1: ncr_ride_bookings.csv

Drop Location (categrica): indica dnde finaliza el viaje; es clave para analizar patrones de demanda segn zonas geogrficas.

Avg VTAT y Avg CTAT (numricas): representan tiempos promedio de turnaround y cycle; tiles para evaluar eficiencia operativa y tiempos de espera.

Booking Value (numrica): refleja el valor econmico del viaje; central para medir ingresos y analizar rentabilidad.

Ride Distance (numrica): distancia recorrida; importante para estudiar relacin con valor, duracin o satisfaccin.

Driver Ratings y Customer Rating (numricas): variables de calidad percibida; fundamentales para medir servicio y satisfaccin.

Payment Method (categrica): permite segmentar usuarios segn forma de pago, lo que aporta a decisiones financieras/comerciales.

 Dataset 2: Sleep_health_and_lifestyle_dataset.csv

Age y Gender (categricas/numricas): permiten segmentar hbitos de sueo y salud segn demografa.

Sleep Duration y Quality of Sleep (numricas): variables centrales para medir calidad del descanso.

Physical Activity Level y Stress Level (numricas): indicadores de estilo de vida, influyen directamente en la salud.

BMI Category (categrica): clasifica peso corporal, til para cruzar con patrones de sueo y actividad.

Blood Pressure, Heart Rate, Daily Steps (numricas/texto): mtricas fisiolgicas que ayudan a vincular sueo con salud cardiovascular.

Sleep Disorder (categrica): variable dependiente clave para analizar factores asociados a trastornos del sueo.

 Dataset 3: CVD_cleaned.csv

General_Health (categrica): autopercepcin de salud, relevante como variable global de estado.

Exercise, Smoking_History, Alcohol_Consumption (categricas/numricas): factores de estilo de vida con alto impacto en riesgo cardiovascular.

Diabetes, Depression, Arthritis, Cancers (categricas): comorbilidades importantes para el anlisis de salud integral.

Height_(cm), Weight_(kg), BMI (numricas): medidas antropomtricas fundamentales para riesgos de CVD (cardiovascular disease).

Fruit_Consumption, Green_Vegetables_Consumption, FriedPotato_Consumption (numricas): reflejan hbitos alimenticios, vinculados a la prevencin o riesgo de CVD.

Heart_Disease (categrica): variable dependiente principal, clave para modelos predictivos o estudios de factores asociados.
